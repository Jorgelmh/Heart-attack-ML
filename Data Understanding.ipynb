## Data Understanding

The purpose of this notebook is to understand the attributes of the dataset and find some errors that could be solve in the data preparation stage.

The dataset contains previous observations of different patients. The aim is to implement a classification model that is able to give accurate predictions as to whether a patient is likely to have a heart attack based on information about their physic health.

The attributes listed in the dataset are:

- Age : Age of the patient


- Sex : Sex of the patient


- exang: exercise induced angina (1 = yes; 0 = no)


- ca: number of major vessels (0-3)


- cp : Chest Pain type chest pain type


    - Value 1: typical angina
    - Value 2: atypical angina
    - Value 3: non-anginal pain
    - Value 4: asymptomatic
    
    
- trtbps : resting blood pressure (in mm Hg)


- chol : cholestoral in mg/dl fetched via BMI sensor


- fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)


- rest_ecg : resting electrocardiographic results


    - Value 0: normal
    - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
    
    
- thalach : maximum heart rate achieved
    

- target : 0= less chance of heart attack 1= more chance of heart attack

```python
n_entries = df.shape[0]

def detectNullValues(data):
    for column in data:
        print("{} | {} | {}".format(
            data[column].name,
            len(data[data[column].isnull()]) / (1.0*n_entries),
            data[column].dtype
        ))
        
detectNullValues(df)
```

## Conclusions of the EDA

- There are no null values on the dataset.
- Most features with continues values have outliers.
- The Age variable seems to have a normal distribution.
- Sex doesn't seem to be correctly distributed, around 2/3 are males (value 1).
- In terms of chest pain, the most recurring scenarios are when people present a typical angina, and in this case there are less likely to have a heart attack than any other chest pain including asymptomatics. People with an atypical angina or non-anginal pain are highly likely to have a heart attack.
- The attribute that represents the resting blood pressure seems to have a normal distribution with a some outliers spotted on the boxplot and the histogram.
- Cholesterol does not seem to have a normal distribution aand has some outliers.
- Most entries have a fasting blood sugar under 120 mg/dl. Note: for those with a FBS greater than 120 mg/dl are 50% likely to have a heart attack.
- There are no visible correlations between continuous variables on the heatmap.
- According to the distribution plot of total heart rate achieved and the output values, people with higher heart rates are more likely to have a heart attack.
- the feature CA that has the number of major vassels must have values between 0-3 according to the documentation, although there entries with value 4.
